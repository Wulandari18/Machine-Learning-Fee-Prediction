```python
# Menyiapkan data training dan data test untuk pemodelan
x_train = dfscale_train
y_train = df_train['Gaji']

x_test = dfscale_test
```

# MODEL KNN

```python
# Memilih model KNN
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV


# Memilih hyperparameter menggunakan GridSearchCV
model = KNeighborsClassifier()
param_grid = {'n_neighbors' : np.arange(2, 50), 'weights' : ['distance', 'uniform']}
gscv = RandomizedSearchCV(model, param_grid, scoring = 'roc_auc', cv = 10)
gscv.fit(x_train, y_train)

gscv.best_params_
```

```python
# Melakukan Cross-Validation
from sklearn.model_selection import cross_val_score

model_knn = KNeighborsClassifier(n_neighbors = 29, weights = 'distance')
cv_value = cross_val_score(model_knn, X_train, y_train, scoring = 'roc_auc', cv = 10)
```

```python
# Nilai score rata-rata 
print('Score rata-rata dari model dengan teknik Cross-Validation:', cv_value.mean())
```

```python
# Memprediksi data test dengan model KNN
y_predict = gscv.predict(x_test)

df_predict = pd.DataFrame(y_predict, columns = ['Gaji'])
df_predict['id'] = df_test['id']

df_predict = df_predict[['id', 'Gaji']]

a = df_predict['Gaji'].value_counts()
print (a)

df_predict.to_csv('newknn_fee_prediction.csv', index = False)
```
